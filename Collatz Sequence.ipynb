```python
import matplotlib.pyplot as plt
T_n = []
N_all = []
for i in range(1, 10001): #starting at the smallest positive integer up to 10000
    N = i #following the loop, N = 1, so we set it back to the next integer
    counter = 0
    while N!=1: #looping N until N = 1
        if N%2==0: #c_n+1 for even
            N //=2
        else: #c_n+1 for odd
            N = 3*N + 1
        counter += 1 #since the loop is set to keep going until N = 1, this counts the length
    T_n.append(counter)
    N_all.append(i) #list of all numbers from 1 to 10000
plt.scatter(N_all, T_n, s = 10)
plt.xlabel("N")
plt.ylabel("T(N)")
```
