```python
import numpy as np
import matplotlib.pyplot as plt
import decimal as Decimal
import fractions as Fraction

x_values = []
y_values = []
for i in range(1,100):
  x = 0.1*i
  x_values.append(x)
  y = (1-np.cos(x))/(x**2)
  y_values.append(y)

plt.plot(x_values, y_values, label = r"$f (x) = \frac{1-cos(x)}{x^2}$")
plt.legend()

x = 1.2e-8
f_x = (1-np.cos(x))/(x**2)
print(f_x) #should have been around 0.5 since 1.2e-8 is near 0, essenntially this value is too high

f_x_new = (2*np.sin(x/2)**2)/(x**2)
print(f_x_new) #yay 0.5 na shea
```
